## 데이터 전처리

### 토큰화

```python
# Okt 토큰화
import konlpy
from konlpy.tag import Okt
okt= Okt()
okt_tokens_list = []
for i in range(len(Feb_1_list)):
    kor_text = str(Feb_1_list[i])
    okt_tokens = okt.morphs(kor_text)
    okt_tokens_list.append(okt_tokens)
print(okt_tokens_list[0])
```

```python
#list in list -> list
okt_tokens_list_total = []
for i in range(len(okt_tokens_list)):
    okt_tokens_list_total.extend(okt_tokens_list[i])
```

```python
type(okt_tokens_list_total)
```

```python
#메모장에 저장
with open('사회1주차_토큰화.txt','w',encoding='UTF-8') as f:
    for t in okt_tokens_list_total:
        f.write(t+'\n')
```

```python
# Okt 품사 태깅
oktTag = []
for token in okt_tokens_list_total:
    oktTag += okt.pos(token)
#print(oktTag)
```

### 1) 전처리

```python
from konlpy.tag import Mecab
mecab = Mecab()

preprocessed_docs = []
for doc in contents :
    # 명사와 동사만으로 문서 전처리
    preprocessed_docs.append(' '.join([token[0] for token in mecab.pos(doc) if token[1][0] in ['N', 'V', '']]))
preprocessed_docs[0][:100]
```

```python
#체언(명사), 용언(동사), 한글이외(외국어, 한자, 숫자)
taggings = ['NNG', 'NNP','NNBC','NG','VV','VA','VX','VCP','VCN','MM','SL','SH','SN']
preprocessed_docs = []
for doc in contents :
    preprocessed_docs.append(' '.join([token[0] for token in mecab.pos(doc) if token[1] in taggings]))
preprocessed_docs[0][:100]
```

### 2) TF-IDF 계산

```python
from sklearn.feature_extraction.text import CountVectorizer

#max_df : 너무 자주 나타나는 단어를 지우기 위함. 0.5 = 50% 이상 나온 단어 무시
count_vectorizer = CountVectorizer(max_df=0.95, min_df=0.01, max_features=10000)
word_count_vector = count_vectorizer.fit_transform(preprocessed_docs)
list(count_vectorizer.vocabulary_.keys())[:10]
```

```python
len(list(count_vectorizer.vocabulary_.keys()))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer(smooth_idf=True, use_idf=True)
tfidf_transformer.fit(word_count_vector)
```

### 3) 핵심키워드 추출

```python
def sort_keywords(keywords):
    return sorted(zip(keywords.col, keywords.data), key=lambda x: (x[1], x[0]), reverse=True)
 
def extract_keywords(feature_names, sorted_keywords, n=5):
    return [(feature_names[idx], score) for idx, score in sorted_keywords[:n]]
```

```python
doc = preprocessed_docs[:] # 핵심키워드 추출할 문서 조회

for i in range(len(preprocessed_docs)):
    feature_names = count_vectorizer.get_feature_names_out() # TF-IDF 단어 목록
    tf_idf_vector = tfidf_transformer.transform(count_vectorizer.transform([preprocessed_docs[i]])) # 문서의 tf-idf 추출
    sorted_keywords += sort_keywords(tf_idf_vector.tocoo()) # TF-IDF를 기준으로 역순 정렬
 
    # 사용자가 지정한 갯수만큼 키워드 추출
keywords = extract_keywords(feature_names, sorted_keywords, 20)
 
print("\n===== 원문 =====")
print(contents[0][:100])
print("\n=== 핵심키워드 ===")
for k in keywords:
    print(k)
```

### 2) 중요 단어 결정

```python
with open('사회1주차_불용어처리완.txt', 'r', encoding='UTF-8') as f:
    words_list = f.readlines()
    for word in words_list:
        print(word)
        print(type(word))
        break
```

```python
print(words_list[:10])
```

```python
# 단어(토큰)의 가중치 계산 및 범위에 포함되는 토큰 식별 
def get_keywords(word_list , min_ratio=0.001, max_ratio=0.5) :
    assert (min_ratio < 1 and max_ratio < 1)
    
    # 토큰별로 빈도수 카운팅
    count_dict = {}    
    for word in word_list:
        count_dict.setdefault(word , 0)
        count_dict[word] +=1
    
    # 분석 문서의 총 토큰수 대비 해당 토큰의 빈도 비율
    keywords = set()
    for word , cnt in count_dict.items():
        word_percentage = count_dict[word] / len (word_list)
        
        # 사전 정의한 비율내에 포함 된 경우 키워드에 추가
        if word_percentage <= max_ratio and word_percentage >=min_ratio:
            keywords.add(word)
    return keywords

get_keywords(words_list)
```
