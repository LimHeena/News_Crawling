```python
news_df['platform'] = news_df['platform'].apply(lambda x: x.replace('Naver', '네이버'))
```

```python
news_df['writed_at'] = news_df['writed_at'].str.replace('오후', 'PM')
news_df['writed_at'] = news_df['writed_at'].str.replace('오전', 'AM')
```

```python
news_df['writed_at'] = pd.to_datetime(news_df['writed_at'], format='%Y.%m.%d. %p %I:%M') #데이트타임으로 변환(형식 맞추기 위해)
```
